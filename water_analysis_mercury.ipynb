```python
# importing packeges
import numpy as np
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt

#import the data from excel sheet
df = pd.read_excel('water sample.xlsx')
fm = pd.DataFrame(df , index = df['Sample']-1)
print('This data for water samples represnts mercury concentraion collected from the river nile \n',fm['Mercury'] ,'\n')
# refering the dataframework
Hg = df['Mercury']
sp = df['Sample']
# statistical analysis for the mercury concentraion 
sta = Hg.describe()
var = round(st.variance(Hg), 5)
range = max(Hg)-min(Hg)
print ('',sta, '\n','The variance = ',var,'\n','The range = ', range)
#visualising 
plt.figure(figsize=(10, 6))
plt.plot(sp ,Hg , marker = 'o')
plt.title('Hg concentraions in the Water samples ')
plt.xlabel('samples')
plt.ylabel('concentration')
plt.grid()
plt.show()
#Box pltting
plt.boxplot(Hg)
plt.legend(Hg)
plt.show()
```
